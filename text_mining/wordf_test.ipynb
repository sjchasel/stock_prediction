```python
def phrase_frequency(s):
        """
        词频统计：计算文章词组频率，利用描述性统计量进行分析（max最大频率，mean平均频率，var频率方差等）
        :return:
        """
        cn_stopwords = deal_wrap(PATH + 'cn_stopwords.txt')
        s = ''.join(split_paragraph(s))
        stopwords = cn_stopwords
        np.append(stopwords,"公司")
        counts = {}
        for ch in '!"#$%&()*+,-./:;<=>?@[\\]^_‘{|}~，。/；‘【】-=《》？：“{}——+、|~·！@#￥%……&*（）1234567890\u3000\xa0 \r\n':
            s = s.replace(ch,"")      # 将文本中特殊字符替换为空格
        for ch in stopwords:
            s = s.replace(ch,"")   
        words = jieba.lcut_for_search(s)     # 搜索引擎模式
        for word in words:
            counts[word] = counts.get(word, 0) + 1    # 遍历所有词语，每出现一次其对应的值加 1
        df = pd.DataFrame({"词":counts.keys(),"词频":counts.values()}).sort_values("词频",ascending=False)
        ans = [df["词频"].mean(),df["词频"].var(),df["词频"].max()]
        return ans
```

```python
def wordss_count(s):
    stopwords =deal_wrap('../词典/cn_stopwords.txt')
    np.append(stopwords,"公司")
    counts = {}
    for ch in '!"#$%&()*+,-./:;<=>?@[\\]^_‘{|}~，。/；‘【】-=《》？：“{}——+、|~·！@#￥%……&*（）1234567890\u3000\xa0 \r\n':
        s = s.replace(ch,"")      # 将文本中特殊字符替换为空格
    for ch in stopwords:
        s = s.replace(ch,"")   
    words = jieba.lcut_for_search(s)     # 搜索引擎模式
    for word in words:
        counts[word] = counts.get(word, 0) + 1    # 遍历所有词语，每出现一次其对应的值加 1
    count_list = list(counts.values())
    ans = [np.mean(count_list),np.var(count_list),np.max(count_list)]
    return ans
```

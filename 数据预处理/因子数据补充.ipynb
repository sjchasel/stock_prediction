```python
import pandas as pd
import numpy as np
gt = pd.read_csv("钢铁行业34支股票.csv")
len(list(gt.columns))
```

```python
newl = ['WAD', 'ADTM',
       'VROC', 'DBCD', 'ATR', 'PRICEOSC', 'BIAS', 'VHF', 'RSI', 'MFI', 'AR',
       'BR', 'ROC', 'SRDM', 'BBI', 'CR', 'PSY']
```

```python
path = 'C:/Users/12968/Desktop/数据科学实战-stock prediction/数据/因子数据补充/钢铁/'
gtcodes = list(gt['code'].unique())
for code in gtcodes:
    name = str(code).zfill(6)+'.csv'
    data = pd.read_csv(path+name)
    for col in list(data.columns):
        print(col)
        if col in newl:
            print(col)
            gt.loc[gt.code==code, col] = data[col]
```

```python
company = pd.read_csv('C:/Users/zhaoyu/Desktop/company.csv',encoding='gbk')
company.drop_duplicates(keep='first',inplace=True)
company.drop(labels=['Unnamed: 0'],axis=1,inplace=True)
```

```python
company1 = company[company['stock_name']=='-']  # 待处理
company2 = company[company['stock_name']!='-']
company1
```

```python
import re
def contain_stocknum(str):
    pattern = re.compile('\([0-9]{6}\)')
    match = pattern.findall(str)
    if match:
        return True
    else:
        return False
    
def code_stock(code):
    agu = pd.read_excel('C:/Users/zhaoyu/Desktop/全部A股代码和名字.xlsx')
    return list(agu[agu['Stkcd'] == int(code)]['Stknme'])[0]
    
for i in range(len(company1)):
    t = company1.iloc[i,8]
    if contain_stocknum(t):
        try:
            company1.iloc[i,7] = re.findall('[0-9]{6}',t)[0]
            company1.iloc[i,6] = code_stock(company1.iloc[i,7])
        except:
            print(t)
        
```

```python
company1 = company1[company1['stock_name']!='-']
```

```python
company = pd.concat([company1,company2],axis=0,ignore_index=True)
company.drop_duplicates(keep='first',inplace=True)
company.to_csv('C:/Users/zhaoyu/Desktop/com.csv',index_label = 'id')
```

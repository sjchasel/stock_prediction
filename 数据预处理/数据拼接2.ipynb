```python
import pandas as pd
import numpy as np
import os
```

```python
com = pd.read_csv('../text_mining/3年3行业的个股研报词句段特征结果.csv')
ind = pd.read_csv('../text_mining/3年3行业的行业研报词句段特征.csv')
zzypath = 'C:/Users/12968/Desktop/数据科学实战-stock prediction/数据/因子数据/种植业/'
gtpath = 'C:/Users/12968/Desktop/数据科学实战-stock prediction/数据/因子数据/钢铁/'
hxylpath = 'C:/Users/12968/Desktop/数据科学实战-stock prediction/数据/因子数据/化学原料/'
```

```python
zzylst = []
gtlst = []
hxyllst = []
for info in os.listdir(zzypath):
    domain = os.path.abspath(zzypath) #获取文件夹的路径
    info = os.path.join(domain,info) #将路径与文件名结合起来就是每个文件的完整路径  
    data = pd.read_csv(info)
    zzylst.append(data)
zzy = pd.concat(zzylst,axis=0,ignore_index=True)
    
for info in os.listdir(gtpath):
    domain = os.path.abspath(gtpath) #获取文件夹的路径
    info = os.path.join(domain,info) #将路径与文件名结合起来就是每个文件的完整路径  
    data = pd.read_csv(info)
    gtlst.append(data)
gt = pd.concat(gtlst,axis=0,ignore_index=True)
    
for info in os.listdir(hxylpath):
    domain = os.path.abspath(hxylpath) #获取文件夹的路径
    info = os.path.join(domain,info) #将路径与文件名结合起来就是每个文件的完整路径  
    data = pd.read_csv(info)
    hxyllst.append(data)
hxyl = pd.concat(hxyllst,axis=0,ignore_index=True)
```

```python
com1 = com[com['ind']=='种植业']
com2 = com[com['ind']=='钢铁']
com3 = com[com['ind']=='化学原料']
zzy.head()
```

```python
com1.columns
```

```python
# 把com1拼到zzy
cols = ['avg_sentence', 'avg_stroke', 'four_word', 'function_ratio', 'n_ratio', 'nine_stroke', 'no_text',
        'sen_per_para', 'sum_para', 'word_max', 'word_mean', 'word_phrase', 'word_var']
for col in cols:
    zzy[col] = 0.000
type(zzy['avg_sentence'][9])
```

```python
#把com1加入zzy
for i in range(len(com1)):
    code = int(com1['stock_num'][i])
    date = com1['date'][i]
    for col in cols:
        #zzy[zzy['date']==date][zzy['code']==code][col] += np.float64(com1[col][i])
        search_se = zzy.loc[zzy['date'] == date, 'code'] == code
        search_index = search_se[search_se == True].index
        zzy[col][search_index] += np.float64(com1[col][i])
        
```

```python
#把com2加入gt
com2.columns
com2 = com2.reset_index(drop = True)
cols = [ 'avg_sentence', 'avg_stroke', 'four_word',
       'function_ratio', 'n_ratio', 'nine_stroke', 'no_text',
       'sen_per_para', 'sum_para', 
       'word_max', 'word_mean', 'word_phrase', 'word_var']
for col in cols:
    gt[col] = 0.000
for i in range(len(com2)):
    code = int(com2['stock_num'][i])
    date = com2['date'][i]
    for col in cols:
        search_se = gt.loc[gt['date'] == date, 'code'] == code
        search_index = search_se[search_se == True].index
        try:
            gt[col][search_index] += np.float64(com2[col][i])
        except:
            gt[col][search_index] += 0
            
```

```python
#把com3加入hxyl
com3.columns
```

```python
com3 = com3.reset_index(drop = True)
cols = [ 'avg_sentence', 'avg_stroke', 'four_word',
       'function_ratio', 'n_ratio', 'nine_stroke', 'no_text',
       'sen_per_para', 'sum_para', 
       'word_max', 'word_mean', 'word_phrase', 'word_var']
for col in cols:
    hxyl[col] = 0.000
for i in range(len(com3)):
    code = int(com3['stock_num'][i])
    date = com3['date'][i]
    for col in cols:
        search_se = hxyl.loc[hxyl['date'] == date, 'code'] == code
        search_index = search_se[search_se == True].index
        try:
            hxyl[col][search_index] += np.float64(com3[col][i])
        except:
            hxyl[col][search_index] += 0
```

```python
ind1 = ind[ind['ind']=='种植业']
ind2 = ind[ind['ind']=='钢铁']
ind3 = ind[ind['ind']=='化学原料']
ind1 = ind1.reset_index(drop = True)
ind2 = ind2.reset_index(drop = True)
ind3 = ind3.reset_index(drop = True)
indlst = [ind1, ind2, ind3]
datalst = [zzy, gt, hxyl]
cols = [ 'avg_sentence', 'avg_stroke', 'four_word',
       'function_ratio', 'n_ratio', 'nine_stroke', 'no_text',
       'sen_per_para', 'sum_para', 
       'word_max', 'word_mean', 'word_phrase', 'word_var']
for ide in range(3):
    industry = indlst[ide]
    data = datalst[ide]
    for i in range(len(industry)):
        date = industry['date'][i]
        for col in cols:
            search_se = data.loc[data['date'] == date]
            search_index = search_se[search_se == True].index
            try:
                data[col][search_index] += np.float64(industry[col][i])
            except:
                data[col][search_index] += 0
```

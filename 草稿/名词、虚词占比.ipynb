```python
import re
import jieba.posseg as peg

fileName = "500个随机样本.txt"
currentFile = open(fileName,encoding='utf-8')
content = currentFile.read()

ratio_n=[]
ratio_pcu=[]
for i in range(500):
    text=re.findall(r'"(.*?)"', content, re.DOTALL)[i]
    words=peg.cut(text) #通过jieba分词获取词的属性
    nw=0
    allw=0
    pcu=0
    for word,flag in words:
        allw=allw+1
        if flag[0]=="n":
            nw=nw+1
        if flag[0]=="p" or flag[0]=="c" or flag[0]=="u":
            pcu=pcu+1
        
    ratio_n.append(nw/allw)
    ratio_pcu.append(pcu/allw)
print(ratio_n)
print(ratio_pcu)
    
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False #用来正常显示负号

plt.hist(ratio_n, bins=30, color='dodgerblue')
plt.title(u"500个随机样本中名词占比分布")
plt.show()
```

```python
plt.hist(ratio_pcu, bins=30, color='lightgreen')
plt.title(u"500个随机样本中虚词占比分布")
plt.show()
```
